```python
from lifelines.plotting import plot_lifetimes
import numpy as np
from numpy.random import uniform, exponential
```

```python
# ä¸€ä¸ªå¸¸è§çš„é”™è¯¯æ˜¯äººä»¬å¯èƒ½åœ¨è¯„ä¼°ç›®æ ‡ç¾¤ä½“å­˜æ´»æ—¶é—´æ—¶ä¼šç®€å•çš„å¿½ç•¥å³ä¾§æˆªæ–­
# è€ƒè™‘ä¸€ç§æƒ…å†µï¼Œå…¶ä¸­äººå£å®é™…ä¸Šç”±ä¸¤ä¸ªäºšç¾¤ç»„æˆï¼Œ ğ´ å’Œ ğµã€‚äººå£ğ´ å¯¿å‘½éå¸¸çŸ­ï¼Œä¾‹å¦‚å¹³å‡2ä¸ªæœˆï¼Œ
# äººå£ ğµ äº«æœ‰æ›´é•¿çš„ä½¿ç”¨å¯¿å‘½ï¼Œä¾‹å¦‚å¹³å‡12ä¸ªæœˆã€‚æˆ‘ä»¬äº‹å…ˆä¸çŸ¥é“è¿™ç§åŒºåˆ«ã€‚å‡å®šæˆ‘ä»¬çš„å®éªŒè¿›è¡Œäº†åä¸ªæœˆ,
# å³åœ¨ğ‘¡=10å°±ç»“æŸäº†ï¼Œæˆ‘ä»¬å¸Œæœ›è°ƒæŸ¥æ•´ä¸ªäººå£çš„å¹³å‡å¯¿å‘½ã€‚
```

```python
N = 25

CURRENT_TIME = 10
```

```python
#æ•°æ®éšæœºç”Ÿæˆã€‚æŒ‡æ•°åˆ†å¸ƒ
actual_lifetimes = np.array([
    exponential(12) if (uniform() < 0.5) else exponential(2) for i in range(N)
])
```

```python
actual_lifetimes
```

```python
#æ•°æ®å³ä¾§æˆªæ–­ï¼Œå¤§äº10ä¸ªæœˆçš„æ•°æ®å½’ç±»æˆ10ã€‚
observed_lifetimes = np.minimum(actual_lifetimes, CURRENT_TIME)
observed_lifetimes
```

```python
#æ•°æ®æ˜¯å¦æ­»äº¡è§‚å¯Ÿæ•°æ®é›†
death_observed = actual_lifetimes < CURRENT_TIME
death_observed
```

```python
ax = plot_lifetimes(observed_lifetimes, event_observed=death_observed)
ax.set_xlim(0, 25)
ax.vlines(10, 0, 30, lw=2, linestyles='--')
ax.set_xlabel("time")
ax.set_title("Births and deaths of our population, at $t=10$")
print("Observed lifetimes at time %d:\n" % (CURRENT_TIME), observed_lifetimes)
```

```python
#æ•°æ®æ²¡æœ‰è¿‡æ»¤ï¼ŒçœŸå®çš„æƒ…å†µï¼Œç”Ÿå­˜åˆ†ææœ€åˆè¢«è®¾è®¡å‡ºæ¥çš„ç›®çš„å°±æ˜¯ä¸ºäº†å¤„ç†å­˜åœ¨å³ä¾§æˆªæ–­çš„è¿™ç§æ•°æ®çš„ã€‚
ax = plot_lifetimes(actual_lifetimes, event_observed=death_observed)
ax.vlines(10, 0, 30, lw=2, linestyles='--')
ax.set_xlim(0, 25)
```
